```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
yr_2016 = "yr_2016.csv"
yr_2017 = "yr_2017.csv"
yr_2018 = "yr_2018.csv"
yr_2019 = "yr_2019.csv"
yr_2020 = "yr_2020.csv"
```

```python
def fixit (file):
    df_year = pd.read_csv(file)
    df_year.starttime = pd.to_datetime(df_year.starttime).dt.floor("Min")
    df_year.stoptime = pd.to_datetime(df_year.stoptime).dt.floor("Min")
    return df_year
```

```python
df_2016 = fixit(yr_2016)
df_2017 = fixit(yr_2017)
df_2018 = fixit(yr_2018)
df_2019 = fixit(yr_2019)
df_2020 = fixit(yr_2020)
```

```python
df_full = pd.concat([df_2016,df_2017,df_2018,df_2019,df_2020],axis = 0)
```

```python
revenue_df = pd.read_csv("revenue_timeline.csv")
revenue_df.columns = revenue_df.columns.str.replace(' ',"_")
```

```python
revenue_df
```

```python
revenue_df.Membership_and_Usage = revenue_df.Membership_and_Usage.str.replace("$","").str.replace(",","")
revenue_df.Sponsorship = revenue_df.Sponsorship.str.replace("$","").str.replace(",","")
revenue_df.Total = revenue_df.Total.str.replace("$","").str.replace(",","")
```

```python
revenue_df.Membership_and_Usage = pd.to_numeric(revenue_df.Membership_and_Usage)
revenue_df.Sponsorship = pd.to_numeric(revenue_df.Sponsorship)
revenue_df.Total = pd.to_numeric(revenue_df.Total)
```
